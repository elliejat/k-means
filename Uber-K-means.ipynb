```python
#Data 
import pandas as pd
import numpy as np

#plots
import seaborn as sns

#dates
from datetime import datetime
```

```python
df = pd.read_csv('yellow_tripdata_2020-01.csv')
```

```python
df.head(5)
```

```python
df.isna().sum()
```

```python
#create total fare column (tips)

df['Total_fare']=df['fare_amount']+ df['tip_amount']
df
```

```python
df2= df[['PULocationID', 'DOLocationID']].copy()
df2
```

```python
x = pd.to_datetime(df['tpep_pickup_datetime'])
y = pd.to_datetime(df['tpep_dropoff_datetime'])
df['duration'] = y-x
```

```python
df
```

```python
#extract and implement k-mean
```

```python
#define
X=df2[['PULocationID', 'DOLocationID']]
y= df.total_amount
X.shape
y.shape
```

```python
from sklearn.cluster import KMeans

kmeans= KMeans(n_clusters= 5, init= 'k-means++', random_state= 42)
y_kmeans= kmeans.fit_predict(X)
y_kmeans
```

```python
#add array to df
df['clusters']=y_kmeans.tolist()
df
```

```python
# linear regression
```

```python
import sklearn.linear_model 
```

```python
X= df[['clusters','duration','trip_distance','total_amount']]
y= df['Total_fare']
```

```python
model= sklearn.linear_model.LinearRegression()

model.fit(X,y)

y_pred= model.predict(X)

print(y_pred)
```

```python
df['Pred_Tot_Fare']=y_pred.tolist()
df
```
